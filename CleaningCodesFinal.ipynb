```python
import pandas as pd  #import the packages
import numpy as np
```

```python
dirty_data = pd.read_excel('youtube.xlsx') #read the file
```

```python
dirty_data.columns = dirty_data.columns.str.title()  #we changed the column names as title
dirty_data.columns
```

```python
dirty_data = dirty_data.rename({'"Id"' : 'Id',
                                '"Yaer"' : 'Year',
                                '"Month"' :'Month',
                                '##Views' : 'Views',
                                '##Likes' : 'Likes',
                                '##Dislikes' : 'Dislikes',
                                '##Comments' : 'Comments',
                                '##Subscribers' : 'Subscribers',
                                '#Video Duration' : 'Video Duration',
                                '"Category"' : 'Category',
                                '"Popularity"' : 'Popularity',
                                '"Rating"' : 'Rating',
                                '"Language"' : 'Language',
                                '"Region"' : 'Region',
                                '"Monetization"' : 'Monetization'}, axis = "columns") #remove the quotation marks
dirty_data.columns  #check the column names after renaming
```

```python
dirty_data.duplicated().sum()   #check duplicates. the answer is 2, so we have drop duplicates
```

```python
dirty_data = dirty_data.drop_duplicates()  #drop duplicates
```

```python
dirty_data.duplicated().sum()    #check if there are any duplicates
```

```python
for i in dirty_data.select_dtypes(include = 'object').columns :   #this code shows us the frequency table of the columns.
    print(dirty_data[i].value_counts())                    # we can see that some columns have the irrelevant punctuations or strings or whitespaces to be removed or replaced.
```

```python
dirty_data['Month'].loc[dirty_data['Month'] == 'August'] = 'Aug'     #replace the August with Aug
dirty_data['Month'].loc[dirty_data['Month'] == 'September'] = 'Sep'    #replace the September with Sep
dirty_data['Month'].value_counts()
```

```python
dirty_data['Category'].loc[dirty_data['Category'] == 'GAMING'] = 'Gaming'   #replace GAMING with Gaming
dirty_data['Category'].loc[dirty_data['Category'] == 'tech'] = 'Tech'        #replace tech with Tech
dirty_data['Category'].value_counts()
```

```python
dirty_data['Language'] = dirty_data['Language'].str.replace("\"","" , regex = True)  #remove the quotation marks
dirty_data['Language'].value_counts()
```

```python
dirty_data['Popularity'] = dirty_data['Popularity'].str.strip()  #remove the whitespaces
dirty_data['Popularity'].value_counts()
```

```python
dirty_data['Video Duration'] = dirty_data['Video Duration'].astype(float)  #convert the data type of the video duration as an float.
                                                        #because video duration column has numerical values
                                                        #but we cannot convert as an float because of the numbers with decimal points
```

```python
dirty_data['Video Duration'] = dirty_data['Video Duration'].astype(float)     #convert data type of the video duration as an float
```

```python
dirty_data.describe().transpose()  #descriptive statistics show us we  do not have any outliers.
```

```python
Q1 = dirty_data['Views'].quantile(0.25)   #define Q1
Q3 = dirty_data['Views'].quantile(0.75)   #define Q2
IQR = Q3 - Q1                 #define IQR
lower_whisker = Q1 - (1.5*IQR)     #define formula for lower whisker
upper_whisker = Q3 + (1.5*IQR)     #define formula for upper whisker
print(dirty_data['Views'].min())
print(lower_whisker)
print(upper_whisker)
print(dirty_data['Views'].max())        #we can see that there is no outlier in views column.
```

```python
Q1 = dirty_data['Likes'].quantile(0.25)    #define Q1
Q3 = dirty_data['Likes'].quantile(0.75)    #define Q3
IQR = Q3 - Q1                       #define IQR
lower_whisker = Q1 - (1.5*IQR)      #define formula for lower whisker
upper_whisker = Q3 + (1.5*IQR)      #define formula for upper whisker
print(dirty_data['Likes'].min())
print(lower_whisker)
print(upper_whisker)
print(dirty_data['Likes'].max())     #we can see that there is no outlier in likes column
```

```python
Q1 = dirty_data['Dislikes'].quantile(0.25)    #define Q1
Q3 = dirty_data['Dislikes'].quantile(0.75)     #define Q2
IQR = Q3 - Q1                                 #define IQR
lower_whisker = Q1 - (1.5*IQR)             #define formula for lower whisker
upper_whisker = Q3 + (1.5*IQR)             #define formula for upper whisker
print(dirty_data['Dislikes'].min())
print(lower_whisker)
print(upper_whisker)
print(dirty_data['Dislikes'].max())         #we can see that there is no outlier in dislikes column
```

```python
Q1 = dirty_data['Comments'].quantile(0.25)      #define Q1
Q3 = dirty_data['Comments'].quantile(0.75)     #define Q3
IQR = Q3 - Q1                               #define IQR
lower_whisker = Q1 - (1.5*IQR)               #define formula for lower whisker
upper_whisker = Q3 + (1.5*IQR)              #define formula for upper whisker
print(dirty_data['Comments'].min())
print(lower_whisker)
print(upper_whisker)
print(dirty_data['Comments'].max())       #we can see that there is no outlier in comments column
```

```python
Q1 = dirty_data['Subscribers'].quantile(0.25)        #define Q1
Q3 = dirty_data['Subscribers'].quantile(0.75)       #define Q3
IQR = Q3 - Q1                                    #define IQR
lower_whisker = Q1 - (1.5*IQR)                  #define formula for lower whisker
upper_whisker = Q3 + (1.5*IQR)                 #define formula for upper whisker
print(dirty_data['Subscribers'].min())
print(lower_whisker)
print(upper_whisker)
print(dirty_data['Subscribers'].max())            #we can see that there is no outlier in subscribers column
```

```python
Q1 = dirty_data['Video Duration'].quantile(0.25)     #define Q1
Q3 = dirty_data['Video Duration'].quantile(0.75)     #define Q3
IQR = Q3 - Q1                                       #define IQR
lower_whisker = Q1 - (1.5*IQR)                     #define formula for lower whisker
upper_whisker = Q3 + (1.5*IQR)                     #define formula for upper whisker
print(dirty_data['Video Duration'].min())
print(lower_whisker)
print(upper_whisker)
print(dirty_data['Video Duration'].max())             #we can see that there is no outlier in video duration column
```

We did not define lower and upper whisker for rating columns. Because we will change the data type of the rating as an object. Because rating has an ordinal variables.

```python
dirty_data['Rating'].loc[dirty_data['Rating'] == -4.6] = 4.6    # replace the unlogical rating values
dirty_data['Rating'].loc[dirty_data['Rating'] == -4.5] = 4.5
dirty_data['Rating'].loc[dirty_data['Rating'] == -4.2] = 4.2
```

```python
dirty_data['Rating'] = dirty_data['Rating'].astype(object)  #turn the data type of the rating as an object because this column has ordinal variables
```

```python
new_df = pd.DataFrame(dirty_data.isnull().sum())
new_df.columns = ['Missing Number']
new_df['Missing Percentage'] = (new_df['Missing Number']*100/dirty_data.shape[0])
new_df                          #the missing table shows us that none of the missing percentages are above 60, that is why there is no need to drop any column
```

```python
dirty_data['Likes'] = dirty_data['Likes'].fillna(dirty_data['Likes'].mean())  #fill null values with the mean because we do not have any outliers.
```

```python
dirty_data['Dislikes'] = dirty_data['Dislikes'].fillna(dirty_data['Dislikes'].mean())  #fill null values with the mean
```

```python
dirty_data['Comments'] = dirty_data['Comments'].fillna(dirty_data['Comments'].mean())  #fill null values with the mean
```

```python
dirty_data['Subscribers'] = dirty_data['Subscribers'].fillna(dirty_data['Subscribers'].mean())  #fill null values with the mean
```

```python
dirty_data['Video Duration'] = dirty_data['Video Duration'].fillna(dirty_data['Video Duration'].mean())  #fill null values with the mean
```

```python
mode_for_category = dirty_data['Category'].value_counts().idxmax()  #create mode value for category
dirty_data['Category'] = dirty_data['Category'].fillna(mode_for_category)  #fill null values with the mode because since it is an object variable
```

```python
mode_for_popularity = dirty_data['Popularity'].value_counts().idxmax()  #create mode value for popularity
dirty_data['Popularity'] = dirty_data['Popularity'].fillna(mode_for_popularity) #fill null values with the mode
```

```python
mode_for_region = dirty_data['Region'].value_counts().idxmax()   #create mode value for region
dirty_data['Region'] = dirty_data['Region'].fillna(mode_for_region)  #fill null values with the mode
```

```python
mode_for_monetization = dirty_data['Monetization'].value_counts().idxmax()  #create mode value for monetization
dirty_data['Monetization'] = dirty_data['Monetization'].fillna(mode_for_monetization)  #fill null values with the mode
```

```python
dirty_data.info()   #check all the data types of the columns are correct
```

```python
dirty_data = dirty_data.drop(['Id', 'Year'], axis = 1)  #drop unnecessary columns
```
