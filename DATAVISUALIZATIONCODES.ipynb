```python
import matplotlib.pyplot as plt     #import the packages
import seaborn as sns
import numpy as np
import pandas as pd
```

```python
youtube_data_clean = pd.read_excel('CLEAN_DATA_FINAL.xlsx')  #read the cleaned dataset
```

```python
plt.figure(figsize= (13,13))    #set the figure size
sns.boxplot(x='Month' , y='Views' , data= youtube_data_clean)  #create a box plot showing the x axis is moonth, y axis is views
plt.xlabel('Month', fontsize=15) #name of the x label
plt.ylabel('Views', fontsize=15) #name of the y label
plt.title('The relationship between Month and Views', fontsize=15 , fontweight='bold')   #title of the plot

```

```python
plt.figure(figsize =(13,13)) #set the figure size
sns.countplot(x='Language' , hue='Monetization', data=youtube_data_clean).legend(bbox_to_anchor=(1.0,1.0))
  #create clustered bar chart which shows the frequency of language with respect to monetization
  #add legend
plt.xticks(rotation=90)  #turn the rotation of the values in the x axis
plt.xlabel('Language' , fontsize =15) #name of the x label
plt.ylabel('Count', fontsize=15)  #name of the y label
plt.title('The Frequency Table of Language with Respect to Monetization', fontsize=15, fontweight='bold')  #title of the plot
```

```python
plt.figure(figsize=(13,13)) #set the figure size
sns.catplot(x='Month' , y='Subscribers', kind='bar' , data=youtube_data_clean, errorbar= None) #create bar plot which shows the Month in the x axis, and the average number of new subscribers gained in the y axis.
plt.ylabel('The Average Number of New Subscribers Gained',fontsize=11)  #the name of the y axis
plt.xlabel('Month', fontsize=11) #the name of the x axis
plt.title('New Subscribers and Month', fontsize=15, fontweight='bold') #title of plot

```

```python
plt.figure(figsize=(13,13)) #set the figure size
sns.violinplot(x='Category', y='Video Duration' , data=youtube_data_clean)  #create a violin plot showing the x axis category, y axis is video duration
plt.ylabel('Video Duration in Seconds', fontsize=15) #name of the y axis
plt.xlabel('Category', fontsize=15) #name of the x axis
plt.title('Relationship between Category and Video Duration', fontsize =15, fontweight='bold') #title of the plot
```

```python
plt.figure(figsize=(13,13)) #set the figure size
sns.boxplot(x='Category' , y='Comments', data=youtube_data_clean)  #create box plot showing the x axis category y axis comment.
plt.xlabel('Category', fontsize=15) #name of the x label
plt.ylabel('Comments' , fontsize =15) #name of the y label
plt.title('The Relationship between Category and Comments',fontsize = 15, fontweight = 'bold') #title of the plot
```
